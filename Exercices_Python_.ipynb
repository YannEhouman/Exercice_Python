#<font color="GREEN"> **Exo 3 : La fonction sum**

Dans cet exercice, la fonction somme prend en entre une liste de nombres et renvoie la somme de ces nombres. Ici, vous n'avez pas le droit d'utiliser la fonction sum de Python. Le but de cet exercice est d'apprendre  implmenter une fonction qui existe dj et gagner en confiance sur nos capacits  faire des choses que d'autres programmeurs plus expriments ont dj fait.

```python
def somme(numbers):
    # Initialiser la variable de somme  0
    total = 0
    # Pour chaque nombre dans la liste
    for number in numbers:
        # Ajouter le nombre  la variable de somme
        total += number
    # Renvoyer la variable de somme
    return total
```

# <font color="PINK">**Exo 8 : Dbut et fin des squences**
En Python, les listes, les tuples et les strings sont des squences. En Python, on peut crire une fonction qui prend diffrents types de variables. Dans cet exercice, nous allons crire une fonction qui prend en Entre, une liste, un tuple ou un string et retourne une variable du mme type mais avec le premier et le dernier lment de la squence.

```python
def premiere_derniere(sequence):
    # Vrifier le type de la squence
    if isinstance(sequence, list):
        # Si c'est une liste, crer une nouvelle liste avec le premier et le dernier lment
        return [sequence[0], sequence[-1]]
    elif isinstance(sequence, tuple):
        # Si c'est un tuple, crer un nouveau tuple avec le premier et le dernier lment
        return (sequence[0], sequence[-1])
    elif isinstance(sequence, str):
        # Si c'est une chane de caractres, crer une nouvelle chane avec le premier et le dernier caractre
        return sequence[0] + sequence[-1]
    else:
        # Si c'est un autre type, lever une exception
        raise ValueError("La squence doit tre une liste, un tuple ou une chane de caractres.")
```

La somme des lments d'index pairs
La somme des lments d'index impairs

```python
def sum_even_odd_indexes(lst):
  even_sum = 0
  odd_sum = 0
  for i in range(len(lst)):
    if i % 2 == 0:
      even_sum += lst[i]
    else:
      odd_sum += lst[i]
  return [even_sum, odd_sum]
```

Ecrivez une fonction qui prend une liste des nombres et qui retourne une liste qui contient la distance entre les diffrents nombres successifs.

```python
def distance_entre_nombres(numbers):
  distances = []
  for i in range(len(numbers) - 1):
    distance = numbers[i+1] - numbers[i]
    distances.append(distance)
  return distances
```

Nous allons crire une fonction generateur_password qui nous gnre des mots de passe. Cette fonction est un peu particulire car elle retournera une autre fonction password. Certains sites n'acceptent que les mots de passe qu'avec des lettres, ou qu'avec des chiffres, ou acceptent des symboles comme @/$ ou simplement un mlange de tout ceci.

La fonction generateur_password prend en entre une chaine de caractres contenant les diffrents caractres possible et retourne la fonction password que je peux xcuter plus tard en lui donnant le nombre de caractres que je veux dans mon mot de passe.

```python
import random

def generateur_password(caracteres_possibles):
  # On dfinit la fonction password  l'intrieur de generateur_password
  def password(longueur):
    # On gnre un mot de passe alatoire en slectionnant alatoirement des caractres
    # dans la chane caracteres_possibles
    pwd = "".join(random.choices(caracteres_possibles, k=longueur))
    return pwd
  # On retourne la fonction password
  return password

# On utilise la fonction generateur_password pour gnrer la fonction password
# qui ne prend en compte que les lettres
password_lettres = generateur_password(string.ascii_letters)
# On utilise la fonction generateur_password pour gnrer la fonction password
# qui ne prend en compte que les chiffres
password_chiffres = generateur_password(string.digits)
# On utilise la fonction generateur_password pour gnrer la fonction password
# qui prend en compte les lettres, les chiffres et les symboles @/$
password_lettres_chiffres_symboles = generateur_password(string.ascii_letters + string.digits + "@/$")

# On utilise les fonctions password gnres pour gnrer des mots de passe de diffrentes longueurs
pwd1 = password_lettres(8)
pwd2 = password_chiffres(12)
pwd3 = password_lettres_chiffres_symboles(16)

print(pwd1)
print(pwd2)
print(pwd3)
```

# <font color="PINK">**Exo 17: Modules**
Dans cet exercice, vous allez cre un module python nomme bootcamp.py qui contient la logique selon laquelle, un tudiant a reussi cette formation.

Ce bootcamp est fait de 4 cours :

La moyenne d'un tudiant est la moyenne coefficie de ses notes. Pour viter que les tudiants ratent des jours de cours, une pnalit est mise en place pour chaque jour de cours rat. Un minus de 5% est fait sur la moyenne de chaque tudiant pour chaque abscence. Un bonus de 10% de la moyenne est accord aux tudiants n'ayant rat aucune sance de cours

Avant de commencer ce bootcamp, chaque tudiant a donn son niveau en Python sur une chelle de 1  5. Pour russir la formation, vous devrez avoir une moyenne finale de :
- 17/20 si votre niveau est 5
- 15/20 si votre niveau est 4
- 14/20 si votre niveau est 3
- 13.5/20 si votre niveau est 2
- 12/20 si votre niveau est 1

Vous tes charg d'crire la logique de russite d'un tudiant afin de faciliter la tche au formateur. Il aimerait entrer les informations de chaque tudiant et obtenir un Oui s'il est admis et Non s'il ne l'est pas.

```python
# bootcamp.py

def reussite_etudiant(niveau, notes, absences):
  # On calcule la moyenne coefficie des notes
  coefficient = [3, 4, 7, 5]
  moyenne = sum(n * c for n, c in zip(notes, coefficient)) / sum(coefficient)
  # On applique la pnalit pour les absences
  moyenne -= (absences * 0.05) * moyenne
  # On applique le bonus pour les tudiants ayant assist  toutes les sances
  if absences == 0:
    moyenne += moyenne * 0.1
  # On dtermine le seuil de russite en fonction du niveau de l'tudiant
  seuils = {
    1: 12,
    2: 13.5,
    3: 14,
    4: 15,
    5: 17
  }
  seuil = seuils[niveau]
  # On retourne True si l'tudiant a russi la formation, False sinon
  return moyenne >= seuil

# Exemples d'utilisation de la fonction reussite_etudiant
print(reussite_etudiant(1, [10, 8, 9, 7], 0))  # True
print(reussite_etudiant(1, [10, 8, 9, 7], 1))  # False
print(reussite_etudiant(3, [10, 8, 9, 7], 2))  # True
print(reussite_etudiant(5, [10, 8, 9, 7], 0))  # False
```

# <font color="LIGHT BLUE">**Exo 18: Programmation oriente objet**
Crez une classe Vehicule avec les attributs suivants:

- couleur
- distance_parcourue (km)
- vitesse maximum (km/h)
- vitesse actuelle (km/h)

On peut donc crer un nouveau vhicule ce cette manire :

- v1 = Vehicule(couleur='bleu', vitesse_maximum='100')
Ajoutez  la classe Vehicule les mthodes ou fonctionalits suivantes:

- accelerer : augmente la vitesse actuelle sur une dure(minutes) sans dpasser la vitesse maximum
- ralentir : diminue la vitesse sur une dure (minutes)
- arreter : elle met  0 la vitesse actuelle

la fonction _repr_ qui permet d'afficher le nombre de distance parcourue et l'tat (arrt, en cours) quand on fait print(v1)

```python
class Vehicule:
  def __init__(self, couleur, vitesse_maximum):
    self.couleur = couleur
    self.distance_parcourue = 0
    self.vitesse_maximum = vitesse_maximum
    self.vitesse_actuelle = 0
  
  def accelerer(self, duree):
    # On calcule la vitesse  laquelle on doit arriver aprs l'acclration
    vitesse_cible = self.vitesse_actuelle + (duree / 60)
    # On ne dpasse pas la vitesse maximum
    self.vitesse_actuelle = min(vitesse_cible, self.vitesse_maximum)
  
  def ralentir(self, duree):
    # On calcule la vitesse  laquelle on doit arriver aprs le ralentissement
    vitesse_cible = self.vitesse_actuelle - (duree / 60)
    # On ne descend pas en dessous de 0
    self.vitesse_actuelle = max(vitesse_cible, 0)
  
  def arreter(self):
    self.vitesse_actuelle = 0
  
  def __repr__(self):
    if self.vitesse_actuelle == 0:
      etat = "arrt"
    else:
      etat = "en cours"
    return f"Distance parcourue : {self.distance_parcourue} km, tat : {etat}"

# Exemples d'utilisation de la classe Vehicule
v1 = Vehicule(couleur='bleu', vitesse_maximum=100)
print(v1)  # Distance parcourue : 0 km, tat : arrt
v1.accelerer(30)
print(v1)  # Distance parcourue : 0 km, tat : en cours
v1.ralentir(60)
print(v1)  # Distance parcourue : 0 km, tat : arrt
v1.arreter()
print(v1)  # Distance parcourue : 0 km, tat : arrt
```

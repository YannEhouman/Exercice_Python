#<font color="red"> **Exo 1 : Devine le nombre**

Ecris une fonction nommée devinette qui ne prend aucun argument. Quand on l'exécute, elle choisit un nombre aléatoire en 0 et 100 (inclus)et demande à l'utilisateur de deviner ce nombre.A chaque fois que l'utilisateur entre un nombre, le programme affiche l'une de ces phrases :

Trop grand
Trop petit
Exact
Si l'utilisateur a choisit correctement, le programme s'arrete. Sinon, le programme demande encore à l'utilisateur de rentrer un nouveau nombre jusqu'à ce qu'il trouve la bonne solution et s'arrête.

```python
import random

def devinette():
  # Choisir un nombre aléatoire entre 0 et 100
  nombre_a_deviner = random.randint(0, 100)

  # Demander à l'utilisateur de deviner le nombre
  devine = int(input("Devine le nombre entre 0 et 100 : "))

  # Boucle qui continue tant que l'utilisateur n'a pas trouvé le bon nombre
  while devine != nombre_a_deviner:
    if devine > nombre_a_deviner:
      print("Trop grand")
    else:
      print("Trop petit")

    devine = int(input("Devine à nouveau : "))

  print("Exact !")
```

#<font color="BLUE"> **Exo 2: Devine le nombre : Niveau 2**
Reprend ton code pour le programme précédent. Modifie le pour donner seulement 3 chances à l'utilisateur pour deviner le nombre aléatoire choisit. Après 3 tentatives echouées, le programme s'arrête et affiche à l'utilisateur qu'il a épuisé ses chances.

```python
import random

def devinette():
  # Choisir un nombre aléatoire entre 0 et 100
  nombre_a_deviner = random.randint(0, 100)

  # Demander à l'utilisateur de deviner le nombre
  devine = int(input("Devine le nombre entre 0 et 100 : "))

  # Compteur de tentatives
  tentatives = 1

  # Boucle qui continue tant que l'utilisateur n'a pas trouvé le bon nombre et n'a pas épuisé ses chances
  while devine != nombre_a_deviner and tentatives < 3:
    if devine > nombre_a_deviner:
      print("Trop grand")
    else:
      print("Trop petit")

    devine = int(input("Devine à nouveau : "))
    tentatives += 1

  # Vérifier si l'utilisateur a trouvé le bon nombre ou s'il a épuisé ses chances
  if devine == nombre_a_deviner:
    print("Exact !")
  else:
    print("Tu as épuisé tes chances. Le nombre à deviner était", nombre_a_deviner)
```

#<font color="GREEN"> **Exo 3 : La fonction sum**

Dans cet exercice, la fonction somme prend en entrée une liste de nombres et renvoie la somme de ces nombres. Ici, vous n'avez pas le droit d'utiliser la fonction sum de Python. Le but de cet exercice est d'apprendre à implémenter une fonction qui existe déjà et gagner en confiance sur nos capacités à faire des choses que d'autres programmeurs plus expérimentés ont déjà fait.

```python
def somme(numbers):
    # Initialiser la variable de somme à 0
    total = 0
    # Pour chaque nombre dans la liste
    for number in numbers:
        # Ajouter le nombre à la variable de somme
        total += number
    # Renvoyer la variable de somme
    return total
```

#<font color="BROWN"> **Exo 5 : Ordonner une chaine de caractères**
Dans cet exercice, vous allez écrire une fonction ordonne_str qui prend une chaîne de caractères (string) et retourne ce même string mais en l'ordonnant par ordre alphabétique.



```python
def ordonne_str(chaine):
    # Convertir la chaîne en une liste de caractères
    liste_caracteres = list(chaine)
    # Trier la liste de caractères par ordre alphabétique
    liste_caracteres_triee = sorted(liste_caracteres)
    # Convertir la liste triée en chaîne de caractères et la renvoyer
    return "".join(liste_caracteres_triee)
```

#<font color="ORANGE"> **Exo 7 : Des phrases sans Voyelle autre que a**
Modifie la précedente fonction pour quelle prenne desormais en entrée une phrase, c'est à dire des mots séparés par des espaces. Il enlève toutes les voyelles sauf la lettre a.


```python
def enleve_voyelles(phrase):
    # Créer une liste des voyelles à enlever
    voyelles = "eiouyEIOUY"
    # Séparer la phrase en mots
    mots = phrase.split()
    # Pour chaque mot
    for i, mot in enumerate(mots):
        # Convertir le mot en une liste de caractères
        liste_caracteres = list(mot)
        # Enlever les voyelles de la liste de caractères
        liste_sans_voyelles = [c for c in liste_caracteres if c not in voyelles]
        # Convertir la liste sans voyelles en chaîne de caractères et la stocker dans la liste de mots
        mots[i] = "".join(liste_sans_voyelles)
    # Rejoindre les mots avec des espaces et renvoyer la phrase
    return " ".join(mots)
```

# <font color="PINK">**Exo 8 : Début et fin des séquences**
En Python, les listes, les tuples et les strings sont des séquences. En Python, on peut écrire une fonction qui prend différents types de variables. Dans cet exercice, nous allons écrire une fonction qui prend en Entrée, une liste, un tuple ou un string et retourne une variable du même type mais avec le premier et le dernier élément de la séquence.

```python
def premiere_derniere(sequence):
    # Vérifier le type de la séquence
    if isinstance(sequence, list):
        # Si c'est une liste, créer une nouvelle liste avec le premier et le dernier élément
        return [sequence[0], sequence[-1]]
    elif isinstance(sequence, tuple):
        # Si c'est un tuple, créer un nouveau tuple avec le premier et le dernier élément
        return (sequence[0], sequence[-1])
    elif isinstance(sequence, str):
        # Si c'est une chaîne de caractères, créer une nouvelle chaîne avec le premier et le dernier caractère
        return sequence[0] + sequence[-1]
    else:
        # Si c'est un autre type, lever une exception
        raise ValueError("La séquence doit être une liste, un tuple ou une chaîne de caractères.")
```

Dans cet exercice, vous allez écrire une fonction qui prend entrée, une liste de nombre et retourne une liste contenant:

La somme des éléments d'index pairs
La somme des éléments d'index impairs

```python
def sum_even_odd_indexes(lst):
  even_sum = 0
  odd_sum = 0
  for i in range(len(lst)):
    if i % 2 == 0:
      even_sum += lst[i]
    else:
      odd_sum += lst[i]
  return [even_sum, odd_sum]
```

# **Exo 11 : Distance entre plusieurs points**
Ecrivez une fonction qui prend une liste des nombres et qui retourne une liste qui contient la distance entre les différents nombres successifs.

```python
def distance_entre_nombres(numbers):
  distances = []
  for i in range(len(numbers) - 1):
    distance = numbers[i+1] - numbers[i]
    distances.append(distance)
  return distances
```

# <font color="ORANGE">**Exo 16: Générateur de mots de passe**
Nous allons écrire une fonction generateur_password qui nous génère des mots de passe. Cette fonction est un peu particulière car elle retournera une autre fonction password. Certains sites n'acceptent que les mots de passe qu'avec des lettres, ou qu'avec des chiffres, ou acceptent des symboles comme @/$ ou simplement un mélange de tout ceci.

La fonction generateur_password prend en entrée une chaine de caractères contenant les différents caractères possible et retourne la fonction password que je peux éxécuter plus tard en lui donnant le nombre de caractères que je veux dans mon mot de passe.

```python
import random

def generateur_password(caracteres_possibles):
  # On définit la fonction password à l'intérieur de generateur_password
  def password(longueur):
    # On génère un mot de passe aléatoire en sélectionnant aléatoirement des caractères
    # dans la chaîne caracteres_possibles
    pwd = "".join(random.choices(caracteres_possibles, k=longueur))
    return pwd
  # On retourne la fonction password
  return password

# On utilise la fonction generateur_password pour générer la fonction password
# qui ne prend en compte que les lettres
password_lettres = generateur_password(string.ascii_letters)
# On utilise la fonction generateur_password pour générer la fonction password
# qui ne prend en compte que les chiffres
password_chiffres = generateur_password(string.digits)
# On utilise la fonction generateur_password pour générer la fonction password
# qui prend en compte les lettres, les chiffres et les symboles @/$
password_lettres_chiffres_symboles = generateur_password(string.ascii_letters + string.digits + "@/$")

# On utilise les fonctions password générées pour générer des mots de passe de différentes longueurs
pwd1 = password_lettres(8)
pwd2 = password_chiffres(12)
pwd3 = password_lettres_chiffres_symboles(16)

print(pwd1)
print(pwd2)
print(pwd3)
```

# <font color="PINK">**Exo 17: Modules**
Dans cet exercice, vous allez crée un module python nommée bootcamp.py qui contient la logique selon laquelle, un étudiant a reussi cette formation.

Ce bootcamp est fait de 4 cours :

- Les bases de Python dont le coefficient est 3
- Le calcul numérique avec Numpy : 4
- Manipulation des données avec Pandas : 7
- Visualisation des données avec Matplotlib et Seaborn : 5
La moyenne d'un étudiant est la moyenne coefficiée de ses notes. Pour éviter que les étudiants ratent des jours de cours, une pénalité est mise en place pour chaque jour de cours raté. Un minus de 5% est fait sur la moyenne de chaque étudiant pour chaque abscence. Un bonus de 10% de la moyenne est accordé aux étudiants n'ayant raté aucune séance de cours

Avant de commencer ce bootcamp, chaque étudiant a donné son niveau en Python sur une échelle de 1 à 5. Pour réussir la formation, vous devrez avoir une moyenne finale de :
- 17/20 si votre niveau est 5
- 15/20 si votre niveau est 4
- 14/20 si votre niveau est 3
- 13.5/20 si votre niveau est 2
- 12/20 si votre niveau est 1

Vous êtes chargé d'écrire la logique de réussite d'un étudiant afin de faciliter la tâche au formateur. Il aimerait entrer les informations de chaque étudiant et obtenir un Oui s'il est admis et Non s'il ne l'est pas.

```python
# bootcamp.py

def reussite_etudiant(niveau, notes, absences):
  # On calcule la moyenne coefficiée des notes
  coefficient = [3, 4, 7, 5]
  moyenne = sum(n * c for n, c in zip(notes, coefficient)) / sum(coefficient)
  # On applique la pénalité pour les absences
  moyenne -= (absences * 0.05) * moyenne
  # On applique le bonus pour les étudiants ayant assisté à toutes les séances
  if absences == 0:
    moyenne += moyenne * 0.1
  # On détermine le seuil de réussite en fonction du niveau de l'étudiant
  seuils = {
    1: 12,
    2: 13.5,
    3: 14,
    4: 15,
    5: 17
  }
  seuil = seuils[niveau]
  # On retourne True si l'étudiant a réussi la formation, False sinon
  return moyenne >= seuil

# Exemples d'utilisation de la fonction reussite_etudiant
print(reussite_etudiant(1, [10, 8, 9, 7], 0))  # True
print(reussite_etudiant(1, [10, 8, 9, 7], 1))  # False
print(reussite_etudiant(3, [10, 8, 9, 7], 2))  # True
print(reussite_etudiant(5, [10, 8, 9, 7], 0))  # False
```

# <font color="LIGHT BLUE">**Exo 18: Programmation orientée objet**
Créez une classe Vehicule avec les attributs suivants:

- couleur
- distance_parcourue (km)
- vitesse maximum (km/h)
- vitesse actuelle (km/h)

On peut donc créer un nouveau véhicule ce cette manière :

- v1 = Vehicule(couleur='bleu', vitesse_maximum='100')
Ajoutez à la classe Vehicule les méthodes ou fonctionalités suivantes:

- accelerer : augmente la vitesse actuelle sur une durée(minutes) sans dépasser la vitesse maximum
- ralentir : diminue la vitesse sur une durée (minutes)
- arreter : elle met à 0 la vitesse actuelle

la fonction _repr_ qui permet d'afficher le nombre de distance parcourue et l'état (arrêt, en cours) quand on fait print(v1)

```python
class Vehicule:
  def __init__(self, couleur, vitesse_maximum):
    self.couleur = couleur
    self.distance_parcourue = 0
    self.vitesse_maximum = vitesse_maximum
    self.vitesse_actuelle = 0
  
  def accelerer(self, duree):
    # On calcule la vitesse à laquelle on doit arriver après l'accélération
    vitesse_cible = self.vitesse_actuelle + (duree / 60)
    # On ne dépasse pas la vitesse maximum
    self.vitesse_actuelle = min(vitesse_cible, self.vitesse_maximum)
  
  def ralentir(self, duree):
    # On calcule la vitesse à laquelle on doit arriver après le ralentissement
    vitesse_cible = self.vitesse_actuelle - (duree / 60)
    # On ne descend pas en dessous de 0
    self.vitesse_actuelle = max(vitesse_cible, 0)
  
  def arreter(self):
    self.vitesse_actuelle = 0
  
  def __repr__(self):
    if self.vitesse_actuelle == 0:
      etat = "arrêt"
    else:
      etat = "en cours"
    return f"Distance parcourue : {self.distance_parcourue} km, état : {etat}"

# Exemples d'utilisation de la classe Vehicule
v1 = Vehicule(couleur='bleu', vitesse_maximum=100)
print(v1)  # Distance parcourue : 0 km, état : arrêt
v1.accelerer(30)
print(v1)  # Distance parcourue : 0 km, état : en cours
v1.ralentir(60)
print(v1)  # Distance parcourue : 0 km, état : arrêt
v1.arreter()
print(v1)  # Distance parcourue : 0 km, état : arrêt
```
